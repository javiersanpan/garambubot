```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("recipes.csv")
df.rename(columns={"Unnamed: 0":"index"}, inplace=True)
df.set_index("index", inplace=True)
df.head(50)
```

```python
import pandas as pd
import ast
import random

# Assuming your DataFrame is already defined as 'df'

def search_recipe_by_ingredient(ingredient):
    matching_recipes = []
    
    for index, row in df.iterrows():
        ingredientes = ast.literal_eval(row['Ingredientes'])
        titulo = row['Titulo']
        elaboracion = row['Elaboracion']
        
        if any(ingredient in ingrediente for ingrediente in ingredientes):
            matching_recipes.append((titulo, ingredientes, elaboracion))
    
    if len(matching_recipes) > 5:
        matching_recipes = random.sample(matching_recipes, 5)
    
    return matching_recipes
```

```python
ingredient = "flor de sabila"
results = search_recipe_by_ingredient(ingredient)

print(results)
```
